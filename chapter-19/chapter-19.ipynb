```python
from numpy import array
from matplotlib import pyplot
data = array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49]
])
print(data)
X, y = data[:,0], data[:,1]
print(X)
print(y)
X = X.reshape((len(X), 1))
print(X)
pyplot.scatter(X, y)
pyplot.show()
```

```python
from numpy import array
from numpy.linalg import inv
from matplotlib import pyplot
data = array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49]
])
print(data)
X, y = data[:,0], data[:,1]
print(X)
print(y)
X = X.reshape((len(X),1))
print(X)
b = inv(X.T.dot(X)).dot(X.T).dot(y)
print(b)
yhat = X.dot(b)
print(yhat)
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```

```python
from numpy import array
from numpy.linalg import inv
from numpy.linalg import qr
from matplotlib import pyplot
data = array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49]
])
print(data)
X, y = data[:,0], data[:,1]
print(X)
print(y)
X = X.reshape((len(X),1))
print(X)
Q, R = qr(X)
print(Q)
print(R)
b = inv(R).dot(Q.T).dot(y)
print(b)
yhat = X.dot(b)
print(yhat)
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```

```python
from numpy import array
from numpy.linalg import lstsq
from matplotlib import pyplot
data = array([
    [0.05, 0.12],
    [0.18, 0.22],
    [0.31, 0.35],
    [0.42, 0.38],
    [0.5, 0.49]
])
print(data)
X, y = data[:,0], data[:,1]
print(X)
print(y)
X = X.reshape((len(X),1))
print(X)
b, residuals, rank, s = lstsq(X, y)
print(b)
print(residuals)
print(rank)
print(s)
yhat = X.dot(b)
print(yhat)
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```
